```python
import numpy as np
import pandas as pd
import datetime
import folium
from tqdm import tqdm
from suntime import Sun, SunTimeException
```

```python
cities = pd.read_excel('cities_sun.xlsx', index_col=0)
cities.head()
```

```python
import folium

m = folium.Map(location=[48, 90], zoom_start=3)

folium.Choropleth(
    geo_data='regions_okato_final.geojson',
    name="choropleth",
    data=cities,
    columns=["okato", "Количество солнечных дней"],
    key_on='feature.properties.okato',
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="",
).add_to(m)
```
